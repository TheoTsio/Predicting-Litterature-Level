```python
import matplotlib.pyplot as plt 


# Plotting Shapley values with color-coded bars
def plot_shapley_values(shapley_values, metric="F1 Score (Weighted)"):
    colors = ['pink' if value >= 0 else 'skyblue' for value in shapley_values.values()]  # Color by sign of Shapley value
    plt.figure(figsize=(8, 6))
    plt.bar(shapley_values.keys(), shapley_values.values(), color=colors)
    plt.xlabel("Groups")
    plt.ylabel(f"Shapley Value ({metric})")
    plt.title(f"Shapley Values for {metric}")
    plt.savefig(r"C:\Users\teots\Downloads\Predicting-Litterature-Level\Results\shaplay_values_of_groups.pdf", format="pdf", bbox_inches="tight")
    plt.show()

plot_shapley_values(shapley_values_f1, metric="F1 Score")
```
